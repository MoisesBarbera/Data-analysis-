#### Then, it is time to train our ML model so that we can find the solution to the problem, in this case, recognise which samples are from benign or malignant tumours.


As the dataframe created contains the solution (target column) we aim to find through our model, we must drop these coulmns to prepare the data properly. 

```python
X = df_cancer.drop(['target'],axis=1)

X  # this dataframe is the same as the one prepared above but only 30 of the 31 columns are present (Target is removed)
```

```python
# As the data from X has the characteristics, then we have to
   # create a Y dataset with the solution for these characteristics,
     # the target column equals, hence, Y
        
y = df_cancer['target']
y
```

```python
# As the data is separated for training we have to devide the data provided for train and test, this way there is
    # a set of data where we train our model and a set of data where we test our model
        
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=5)

# Here we agree the size of the test sample to be 20% 
 # so we train our model in the rest 80% 
```

We can observe the size of the different samples for each action here:

```python
X_train.shape     # Train shape for characteristics data
```

```python
X_test.shape      # Test shape for characteristics data
```

```python
y_train.shape     # Train shape for Target/answer data
```

```python
y_test.shape      # Test shape for Target/answer data
```

Finally, to train the data now that it has been properly divided we use the following libraries from sklearn:

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix

# And we create the model in this section

svc_model = SVC()
svc_model.fit(X_train, y_train)
```

#### After the model has been created we start evaluating it in a quantitative and visual apprach. Based on our results we consider improving the model or not.


To evaluate, we create a Y prediction from the X_test data. If the model has been properly created, Y prediction sshould agree with Y_test only that this time, the decision have been made by the model rather than assigned by us.

Then the visualization can be done using a confusion matrix approach.

```python
y_predict = svc_model.predict(X_test)      # Creates Y_prediction
cm = confusion_matrix(y_test, y_predict)   # Compares predicted and real Y
```

```python
sns.heatmap(cm, annot=True) # displays the confusion matrix annotating the right ang wrong guesse
```

Remember that 0 is malign and 1 is benign. In the confusion matrix the X axis show the True Class/values and the Y axis show the predictions.

If our model classifies data in (0, 0) we are certain it is a malignant tumour, if it assigns data in (1, 1) we are certain it is a benign tumour.

On the other hand if there is data in (1, 0) we have our model predicting a malignant tumour when in fact is a benign one [Type 1 error] and the data in (0, 1) would correspond to our model predicting benign when it was actually malignant [Type 2 error].

```python
print(classification_report(y_test, y_predict))  # prints a report of the obtained data
```

#### As seen, the evaluation of the model returns a small level of accuracy so we should proceed to improve the model and re-evaluate

To improve the model we scale it doing a unity-based normalization where we bring all values into a range [0, 1].

```python
min_train = X_train.min()  # We evaluate the minimum values from train set X
min_train
```

```python
range_train = (X_train - min_train).max()  # here we create the range of action for our normalization
range_train 
```

```python
X_train_scaled = (X_train - min_train)/range_train  # We normalise the X_train data
X_train_scaled
```

```python
# Before normalisation, the data from mean values were spread in a 0 to 2500 range
# Making more difficult the assignation of a boundary.

sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
```

```python
# After normalisation, now with the range 0 to 1, makes easier the assignation of the barrier

sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
```

```python
# Here we repeat the normalisation process from train data for test data

min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

```python
# using the previous loaded skearn libraries we generate the new svc model with the fitted data

svc_model = SVC()
svc_model.fit(X_train_scaled, y_train)
```

```python
# Generate new prediction data and visualizatin of results

y_predict = svc_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_predict)

sns.heatmap(cm,annot=True,fmt="d")
```

```python
print(classification_report(y_test,y_predict))
```

#### After normalisation the model has improved accuracy by 95.4% to 96%. But perhaps the model could be improve a bit more.

```python
# To improve we create a new model with better deffined parameters
# we use param_grid to find the best parameters for our model
# C: assigns the new ranges to look for
# gamma created the grid 
# kernel: radial basis function

param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']} 
```

```python
# we import the GridSearchCV library to create the new grid using the parametres described above

from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
# Now that the new grid with the ML model has been created we fit the train data

grid.fit(X_train_scaled,y_train)
```

```python
# The best parameters from this fit applied on the grid are:

grid.best_params_
```

```python
# And the best estimator:

grid.best_estimator_
```

```python
# We prepare again the predictions for the target data (Y) as done previously

grid_predictions = grid.predict(X_test_scaled)
```

```python
# We create the new confusion matrix

cm = confusion_matrix(y_test, grid_predictions)
```

```python
# Finally we visualise the results from our predictive ML model

sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,grid_predictions))
```

In this study I have imported data from Skleran datasets about Breast Cancer diagnosis, recording 569 diagnosis contining the 30 key characteristics to assess a kind of cancer as in benign or malignant.
